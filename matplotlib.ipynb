```python
import matplotlib.pyplot as p
import numpy as np
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
p.plot(x,y)
```

```python
p.plot(x,y,color='pink',linewidth=5, linestyle='dotted')
```

```python
p.xlabel('Day')
p.ylabel('Temperature')
p.title('Weather')
p.plot(x,y,color='green',linewidth=5, linestyle='dashdot') #change plot design using plot api
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
p.xlabel('Days')
p.ylabel('Temperature')
p.title('Weather')
p.plot(days,max_t, label = 'Tmax')
p.plot(days, min_t, label = 'Tmin')
p.plot(days,avg_t, label = 'Tavg') 
p.legend()
p.grid()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
p.xlabel('Company Name')
p.ylabel('Revenue in Billions')
p.title('US revenue')
p.bar(company,revenue, label='revenue')
p.legend()
```

```python
profit=[40,2,34,12]
```

```python
p.bar(company,revenue, label='revenue',width=0.4)
p.bar(company,profit, label='profit',width=0.4)
p.legend()
```

```python
xpos = np.arange(len(company))
p.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
p.bar(xpos+0.2,profit, width=0.4,label="Profit")

p.xticks(xpos,company)
p.ylabel("Revenue(Bln)")
p.title('US Technology Stocks')
p.legend()
```

HISTOGRAMS

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
p.hist(blood_sugar, bins=3, rwidth = .95)
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

p.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
p.legend()
```

PIE CHART

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]

p.axis('equal')
p.pie(exp_vals, labels = exp_labels, autopct='%0.1f%%')
p.show()
```
